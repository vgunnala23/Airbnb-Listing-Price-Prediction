```

from pyspark.sql.functions import *

df = df.withColumn('Description', regexp_replace('Description', '0', ''))
df = df.withColumn('Description', regexp_replace('Description', '#N/A', ''))
```

```
from pyspark.ml.feature import Tokenizer, StopWordsRemover
from nltk.stem.snowball import SnowballStemmer
from pyspark.sql.functions import udf, col, lower, regexp_replace
from pyspark.sql.types import *

# Creating a df with numerical and categorical columns and cleaning the text data
df_clean = df.select('id','number_of_reviews','price','neighbourhood','latitude','longitude','room_type','minimum_nights','Calculated_host_listings_count',(lower(regexp_replace('Description', "[^a-zA-Z\\s]", "")).alias('Description')))

# Tokenizing text data
tokenizer = Tokenizer(inputCol='Description', outputCol='words_token')
df_words_token = tokenizer.transform(df_clean).select('id','number_of_reviews','price','neighbourhood','latitude','longitude', 'room_type','minimum_nights', 'Calculated_host_listings_count','words_token')

# Removing stop words
remover = StopWordsRemover(inputCol='words_token', outputCol='words_clean')
df_words_no_stopw = remover.transform(df_words_token).select('id','number_of_reviews','price','neighbourhood','latitude','longitude', 'room_type','minimum_nights', 'Calculated_host_listings_count','words_clean')

# Stemming the words to canonical form
stemmer = SnowballStemmer(language='english')
stemmer_udf = udf(lambda tokens: [stemmer.stem(token) for token in tokens], ArrayType(StringType()))
df_stemmed = df_words_no_stopw.withColumn("words_stemmed", stemmer_udf("words_clean")).select('id','number_of_reviews','price','neighbourhood','latitude','longitude', 'room_type','minimum_nights','Calculated_host_listings_count', 'words_stemmed')

# Filtering words with length > 3
filter_length_udf = udf(lambda row: [x for x in row if len(x) >= 3], ArrayType(StringType()))
df_final_words = df_stemmed.withColumn('words', filter_length_udf(col('words_stemmed')))
```

```
df_final = df_final_words.dropna(how='any')
```

```
df_final.count()
```

```
(train_set, test_set) = df_final.randomSplit([0.80, 0.20], seed = 23)
```

```
from pyspark.ml.feature import HashingTF, IDF, Tokenizer
from pyspark.ml.feature import VectorAssembler,StringIndexer
from pyspark.ml import Pipeline

# Generating tf-idf from the text data 

hashtf = HashingTF(numFeatures=2**16, inputCol="words", outputCol='tf')
idf = IDF(inputCol='tf', outputCol="tfidf", minDocFreq=10) #minDocFreq: remove sparse terms

# Indexing the Categorical variable

neighbourhood_indexer = StringIndexer(inputCol='neighbourhood',outputCol='neighbourhood_index',handleInvalid='keep')
roomType_indexer = StringIndexer(inputCol='room_type',outputCol='room_type_index',handleInvalid='keep')

# Vector assembler is used to create a vector of input features

assembler = VectorAssembler(inputCols=['tfidf','number_of_reviews','room_type_index','neighbourhood_index', 'latitude','longitude', 'Calculated_host_listings_count'], outputCol="features")


pipeline = Pipeline(stages=[hashtf, idf, roomType_indexer, neighbourhood_indexer, assembler])

pipelineFit = pipeline.fit(train_set)

# Transforming the train set
train_data = pipelineFit.transform(train_set)

# Transforming the test set
test_data = pipelineFit.transform(test_set)

train_data.show(5)
```

```
from pyspark.ml.regression import LinearRegression

lr_model = LinearRegression(labelCol='price')

fit_model = lr_model.fit(train_data.select(['features','price']))
```

```
results = fit_model.transform(test_data)
```

```
results.select(['price','prediction']).show()
```

```
trainingSummary = fit_model.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```

```
# The linear regression model has performed relatively better using the textual data from the 'description' column of the data. The RMSE suggests that the average prediction amount of price is off by $110, which can be certainly improved using transformations of some parameters and including derived variables. The R-Squared value confirms that the variables used in the model were able to explain 50% of variation in the price. 

#Overall we can say that including textual data in our model did improve the results although very minimally but did open up the scope further improvement. 
```
